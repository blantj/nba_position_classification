# Scrape Data

Create Selenium Webdriver

```python
driver_path = '/Users/blantj/documents/Practice/ncaa_position_clustering/chromedriver'
options = Options()
options.headless = True
options.add_argument("--window-size=1920,1200")
driver = webdriver.Chrome(options=options, executable_path=driver_path)
```

Scrape list of ncaa team urls

```python
driver_url = 'https://www.sports-reference.com/cbb/seasons/2019-standings.html'
driver.get(driver_url)
conference_sizes = [12, 15, 9, 14, 9, 10, 10, 11, 11, 14, 9, 10, 14, 10, 8, 11, 13, 12, 10, 11, 10, 12, 12, 10, 14,
                   10, 13, 9, 12, 10, 9, 10]
urls = []
for conf in range(2,len(conference_sizes)+2):
    for team in range(1,conference_sizes[conf-2]+1):
        if f'/html/body/div[2]/div[5]/div[{conf}]/div[4]/div/table/tbody/tr[{team}]/td[1]/a' != "/html/body/div[2]/div[5]/div[18]/div[4]/div/table/tbody/tr[7]/td[1]/a":
            urls.append(driver.find_element_by_xpath(f'/html/body/div[2]/div[5]/div[{conf}]/div[4]/div/table/tbody/tr[{team}]/td[1]/a')
            .get_attribute('href'))
```

Create players and stats lists

```python
players = []
stats = []
```

Scrape player position and stats by team

```python
count = 313
for url in urls[count:]:
    team = url.split('/')[-2]
    driver.get(url)
    for player in range(1,20):
        try:
            players.append([
              driver.find_element_by_xpath(f'/html/body/div[2]/div[5]/div[3]/div[2]/div/table/tbody/tr[{player}]/th/a').text
            , driver.find_element_by_xpath(f'/html/body/div[2]/div[5]/div[3]/div[2]/div/table/tbody/tr[{player}]/td[3]').text
            , team])
            stats.append([
              driver.find_element_by_xpath(f'/html/body/div[2]/div[5]/div[7]/div[2]/div/table/tbody/tr[{player}]/td[1]/a').text
            , team
            , driver.find_element_by_xpath(f'/html/body/div[2]/div[5]/div[7]/div[2]/div/table/tbody/tr[{player}]/td[2]').text
            , driver.find_element_by_xpath(f'/html/body/div[2]/div[5]/div[7]/div[2]/div/table/tbody/tr[{player}]/td[3]').text
            , driver.find_element_by_xpath(f'/html/body/div[2]/div[5]/div[7]/div[2]/div/table/tbody/tr[{player}]/td[4]').text
            , driver.find_element_by_xpath(f'/html/body/div[2]/div[5]/div[7]/div[2]/div/table/tbody/tr[{player}]/td[5]').text
            , driver.find_element_by_xpath(f'/html/body/div[2]/div[5]/div[7]/div[2]/div/table/tbody/tr[{player}]/td[6]').text
            , driver.find_element_by_xpath(f'/html/body/div[2]/div[5]/div[7]/div[2]/div/table/tbody/tr[{player}]/td[7]').text
            , driver.find_element_by_xpath(f'/html/body/div[2]/div[5]/div[7]/div[2]/div/table/tbody/tr[{player}]/td[8]').text
            , driver.find_element_by_xpath(f'/html/body/div[2]/div[5]/div[7]/div[2]/div/table/tbody/tr[{player}]/td[9]').text
            , driver.find_element_by_xpath(f'/html/body/div[2]/div[5]/div[7]/div[2]/div/table/tbody/tr[{player}]/td[10]').text
            , driver.find_element_by_xpath(f'/html/body/div[2]/div[5]/div[7]/div[2]/div/table/tbody/tr[{player}]/td[11]').text
            , driver.find_element_by_xpath(f'/html/body/div[2]/div[5]/div[7]/div[2]/div/table/tbody/tr[{player}]/td[12]').text
            , driver.find_element_by_xpath(f'/html/body/div[2]/div[5]/div[7]/div[2]/div/table/tbody/tr[{player}]/td[13]').text
            , driver.find_element_by_xpath(f'/html/body/div[2]/div[5]/div[7]/div[2]/div/table/tbody/tr[{player}]/td[14]').text
            , driver.find_element_by_xpath(f'/html/body/div[2]/div[5]/div[7]/div[2]/div/table/tbody/tr[{player}]/td[15]').text
            , driver.find_element_by_xpath(f'/html/body/div[2]/div[5]/div[7]/div[2]/div/table/tbody/tr[{player}]/td[16]').text
            , driver.find_element_by_xpath(f'/html/body/div[2]/div[5]/div[7]/div[2]/div/table/tbody/tr[{player}]/td[17]').text
            , driver.find_element_by_xpath(f'/html/body/div[2]/div[5]/div[7]/div[2]/div/table/tbody/tr[{player}]/td[18]').text
            , driver.find_element_by_xpath(f'/html/body/div[2]/div[5]/div[7]/div[2]/div/table/tbody/tr[{player}]/td[19]').text
            , driver.find_element_by_xpath(f'/html/body/div[2]/div[5]/div[7]/div[2]/div/table/tbody/tr[{player}]/td[20]').text
            , driver.find_element_by_xpath(f'/html/body/div[2]/div[5]/div[7]/div[2]/div/table/tbody/tr[{player}]/td[21]').text
            , driver.find_element_by_xpath(f'/html/body/div[2]/div[5]/div[7]/div[2]/div/table/tbody/tr[{player}]/td[22]').text
            , driver.find_element_by_xpath(f'/html/body/div[2]/div[5]/div[7]/div[2]/div/table/tbody/tr[{player}]/td[23]').text
            , driver.find_element_by_xpath(f'/html/body/div[2]/div[5]/div[7]/div[2]/div/table/tbody/tr[{player}]/td[24]').text
            , driver.find_element_by_xpath(f'/html/body/div[2]/div[5]/div[7]/div[2]/div/table/tbody/tr[{player}]/td[25]').text])
        except:
            pass
    count+=1
    
```
